# PreProcessing of DataSet Begins

```python
#from the statistical view, lights have an insignificant impact with zero entries seen in the interquartile range
#further confirming this by evaluating the distribution of values within the column
df.lights.value_counts()
```

```python
#With 15252 zero entries (light columns) in 19735 rows, the lights column will add no value to the model
# Hence proceeding to drop column
#Also date column will be dropped as this is not a time series problem
df = df.drop(["date","lights"], axis = 1)
df.head(2)
```

```python
#it is important our dataset is normalized to a common scale to ensure a good interpretation. this is done by using the min max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
#Creating a Dataframe for the scaled df
new_df = pd.DataFrame(scaler.fit_transform(df), columns =df.columns)

#seperating predictor and response variables
pred_vars = new_df.drop(columns =['Appliances'])
response_var = new_df['Appliances']
```

```python
new_df.corr()
```

# QUESTION 12 : FIT A LINEAR MODEL BETWEEN T2 AND T6. WHAT IS THE R^2

```python
train_x = df[['T2']]
train_y = df[['T6']]

#  splitting our dataset into the training and testing 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_x, train_y, test_size = 0.3, random_state = 42)

#importing linear regression
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

#fitting the model to the training dataset
linear_model.fit(x_train, y_train)

#obtaining predictions
predicted_values = linear_model.predict(x_test)


#R-Squared(Coefficient of Determination)
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 3)
```

# QUESTION 13: FIND THE MEA

```python
#importing linear regression
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

# now, splitting our dataset into the training and testing 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(pred_vars, response_var, test_size = 0.3, random_state = 42)

#fitting the model to the training dataset
linear_model.fit(x_train, y_train)

#obtaining predictions
predicted_values = linear_model.predict(x_test)

```

```python
#evaluating regression performance using appropriate metrics
#Mean Absolute Error(MAE)
from sklearn.metrics import mean_absolute_error
mae =mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

# QUESTION 14 :FIND THE RSS

```python
#Residual sum of squares(RSS)
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 3)
```

# QUESTION 15: FIND THE RMSE

```python
##Root mean squared error(RMSE)
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test , predicted_values))
round(rmse, 3)
```

# QUESTION 18: TRAIN A RIDGE MODEL WITH AN ALPHA OF 0.4 AND GET THE RMSE

```python
#ridge regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
#fitting the model to the training dataset
ridge_reg.fit(x_train, y_train)
#obtaining predictions
predicted_reg = ridge_reg.predict(x_test)
```

# QUESTION 19: TRAIN A LASSO MODEL HAVING AN ALPHA OF 0.001 AND OBTAIN THE NEW FEATURE WEIGHTS

```python
#feature selection and lasso regression
from  sklearn.linear_model import Lasso
#defining the model and alpha
lasso_reg = Lasso(alpha = 0.001)
#fitting the model to the training dataset
lasso_reg.fit(x_train, y_train)
#obtaining predictions
predicted_lasso = lasso_reg.predict(x_test)

feature_weights = lasso_reg.coef_

weights_value = []
for i in feature_weights:
    if i !=0:
        weights_value.append(i)
weight_value_count = len(weights_value)
weight_value_count
```
